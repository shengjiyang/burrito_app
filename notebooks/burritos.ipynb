```python
import pandas as pd

df = pd.read_csv("data/burritos_01022018.csv")

print(df.shape)
df.head()
```

```python
df.columns
```

```python
count = df.Burrito.value_counts().sort_values(ascending=True)
count
```

```python
types = list(count.to_dict().keys())
types
```

```python
import matplotlib.pyplot as plt

f, ax = plt.subplots(figsize=(20, 17))

plt.barh(types, count.to_list());
```

```python
print("NaN Counts")
print("----------\n")

for col in df.columns:
    print(col)
    print("-----------")
    print(df[col].isna().sum())
```

```python
simplified = df[["Volume", "Tortilla", "Temp", "Meat", "Fillings", "Meat:filling", "Uniformity", "Salsa", "Synergy", "Wrap"]]

print(simplified.shape)
simplified.head()
```

```python
simplified.Volume.describe()
```

```python
import numpy as np

avg_vol = simplified.Volume.describe().to_dict()["mean"]

for i in range(len(df.Volume.values)):
    if df.Volume[i] == np.nan:
        df.Volume[i] = avg_vol
        
```
